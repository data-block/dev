# 인사 데이터 - 이직 분류 모형
### 이직 여부에 따른 변수 특성 분포 파악 (EDA)
- satisfaction_level 직무만족도
- last_evaluation : 최근 평가
- number_project : 맡은 프로젝트 수
- average_montly_hours : 월 평균 근로시간
- time_spend_company : 근속년수
- Work_accident: 사건/사고의 유무
- left : 이직 유무 (1 True, 0 False)
- promotion_last_5years :5년 사이 진급 유무
- sales : 부서
- salary ; 연봉

```r
data2 <- read.csv('../data/HR_comma_sep.csv')
str(data2)
head(data2)
```

```r
library(ggplot2)
library(plyr)

GRAPH1=ddply(data2,.(salary,sales),summarise,
             MEAN=mean(satisfaction_level),MAX=max(satisfaction_level))
```

```r
head(GRAPH1)
```

```r
ggplot(GRAPH1)+
  geom_tile(aes(x=salary,y=sales,fill=MEAN))+
  scale_fill_gradientn(colours = c("white", "green", "red"), values = c(0,0.1,1))
```

```r
ggplot(GRAPH1)+
  geom_point(aes(x=salary,y=sales,col=MEAN,size=MEAN))+
  scale_color_gradientn(colours = c("white", "royalblue", "red"), values = c(0,0.1,1))
```

# 카이제곱 (독립성) 검정
- 중요한 이유
    - 실제 데이터에서는 대부분 범주형이다.
    - 그래서 카이제곱이 중요하게 쓰인다.
- EDA 성격
    - 모델링은 아니고, 두 (범주형) 변수들의 간의 연관관계를 보기 위한 것
    - 결과가 아니라 과정
        - 흑인, 백인 범죄에 따른 사형 집행을 통해 알아보는 인종차별
        - 지역별 정당 지지도

    - 기각이 안 됐으면, 두 변수는 독립 ... 인종 차별이 없다.
    - 기각이 됐으면, 독립 X
    - 모델링: x가 변하면, Y도 변한다.
- 현황파악!
- 차이를 보는 분석에 속한다. 엄밀히 따지면 관계를 보는 분석은 아니다.
- 아노바는 평균차이가 0인지 아닌지 파악하는데, 범주형은 그런 게 아니다. 그래서 이론과 현실로 비교
