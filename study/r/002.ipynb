## Basic Manipulation
### 1. rename column

```r
Carseats <- rename(Carseats, Sales = Revenue)          # dplyr
colnames(Carseats)
```

```r
names(Carseats)[names(Carseats)=="Sales"] <- "Revenue" # base
colnames(Carseats)
```

###  2-1. filter 1 - Income

```r
temp <- filter(Carseats, Income > 100)    # dplyr
temp <- Carseats %>% filter(Income < 100) # dplyr
temp <- Carseats[Carseats$Income > 100, ] # base
```

### 2-2. filter 2 - Age

```r
temp <- filter(Carseats, Age >= 30 & Age < 40)                   # dplyr
temp <- Carseats %>% filter(Age >= 30 & Age < 40)                # dplyr
temp <- Carseats[((Carseats$Age >= 30) & (Carseats$Age < 40)), ] # base
```

### 3. select

```r
temp <- select(Carseats, Income, Population)     # dplyr
temp <- Carseats %>% select(Income, Population)  # dplyr
temp <- Carseats[, c("Income", "Population")]    # base
```

### 4. arrange

```r
Carseats <- arrange(Carseats, Price)         # dplyr
Carseats <- Carseats %>% arrange(Price)      # dplyr
Carseats <- Carseats[order(Carseats$Price),] # base
```

```r
Carseats <- arrange(Carseats, desc(Price))
Carseats <- Carseats %>% arrange(desc(Price))
Carseats <- Carseats[order(Carseats$Price, decreasing = TRUE), ]
```
