** 10. 9번 데이터에서 현재 체중과 측정 체중이 가장 차이가 크게 나는 순으로 정렬해주세요. **

```python
# Write your code here!
```

```python
print(weight_member["현재 체중"].dtypes)
print(weight_member["측정 체중"].dtypes)
```

```python
weight_member["현재 체중"].value_counts()
```

```python
def convert_weight_to_integer(weight):
    if pd.isnull(weight):
        return weight
    
    if "kg" in weight:
        weight = weight.replace("kg", "")
        
        return int(weight)
    
    return int(weight)


weight_member["Current Weight(int)"] = weight_member["현재 체중"].apply(convert_weight_to_integer)

print(weight_member.shape)
weight_member.head()
```

```python
weight_member["loss"] = weight_member["Current Weight(int)"] - weight_member["측정 체중"]

print("[loss] min = {0}, max = {1}, mean = {2:.2f}".format(weight_member["loss"].min(), weight_member["loss"].max(), weight_member["loss"].mean()))

print(weight_member.shape)
weight_member[["Current Weight(int)", "측정 체중", "loss"]].head()
```

```python
weight_member['loss'].value_counts()
```
