
### 2. 구구단을 아래와 같이 출력하는 함수 times_table(k)를 작성하세요.

- k : 출력할 구구단의 숫자가 저장된 리스트 예: [2,4,5]

<pre>

````python
###################################################
# pass로 표시된 곳에 적절히 코드를 써 넣으세요.
###################################################

def times_table(K) :
    # 구구단에서 뒤쪽 숫자 1~9까지 루프를 돌면서
    for pass :
        row = ""
        
        # 구구단에서 k단에 대해서
        for k in K :
            # 출력의 한 행을 만든다.
            row += pass
            
            #구구단 사이의 빈 간격
            row += "   " 
            
        # 한줄 다 만들었으면 출력한다.    
        print(row)
            
times_table([2,3,4,5,6,7,8,9])        
````

### 5. 종합 문제 K-mean clustering

````python
import pandas as pd

plt.rcParams["figure.figsize"] = (15,5)

faithful = pd.read_csv('faithful.csv')
faithful.head()

faithful = (faithful - faithful.mean()) / faithful.std()
faithful.columns = ['eruptions', 'waiting']

fig, (ax1, ax2) = plt.subplots(1, 2)

ax1.scatter(faithful.eruptions, faithful.waiting)
ax1.set_title('Old Faithful Data Scatterplot')
ax1.set_xlabel('Length of eruption (minutes)')
ax1.set_ylabel('Time between eruptions (minutes)')

centroids = np.array([[-1.3, 1.0], [1.2, -1.0]])
ax2.scatter(faithful.eruptions, faithful.waiting)
ax2.plot(centroids[0,0], centroids[0,1], 'bx', markersize=10, markeredgewidth=5)
ax2.plot(centroids[1,0], centroids[1,1], 'rx', markersize=10, markeredgewidth=5)
plt.show()

faith = np.array(faithful)
k = 2

print(faith.shape)
````

- 위 그림은 미국 Yellowstone 국립 공원의 유명한 간헐천(間歇泉)의 분출 지속시간과 분출간의 시간차를 나타낸 것이다.

- http://www.stat.cmu.edu/~larry/all-of-statistics/=data/faithful.dat

- 그래프상 분출과 분출사이의 간격이 길수록 오래동안 분출하는 것처럼 보인다.

- 분출사이의 간격과 분출 지속시간을 두 그룹으로 묶어볼 수 있을 것 같다.

- 오른쪽 그래프처럼 두 묶음의 중심(파란색, 빨간색)을 아무렇게나 찍은 다음 다음 절차를 통해 묶음 작업을 진행하고자 한다.

    - 1) 그림처럼 초기 중심점을 2개 아무렇게나 찍는다.
    - 2) 모든 데이터에 대해서 초기 중심점과의 간격을 계산하여 더 가까운쪽의 묶음으로 데이터를 할당한다.
    - 3) 위 작업이 완료되면 두 묶음으로 분리된 데이터의 평균점을 구한다.
    - 4) 계산된 평균점을 새로운 점으로 하여 2의 과정으로 돌아가 반복한다.

- 위 같은 반복을 5회한 후 다음과 같은 그림을 얻게 된다.

<img src="imgs/hw1.png"/>

- 이 과정은 K-mean clustering 알고리즘으로 알려져있고 구현은 다음과 같습니다.

- 주석을 참고하여 pass라고 되어 있는 곳에 적당한 코드를 완성하세요.

````python
# 주의 : 이 셀을 실행시키기 전에 반드시 윗 셀을 먼저 실행시켜 데이터를 로드하고 실행하여야 함. 

###################################################
# pass로 표시된 곳에 적절히 코드를 써 넣으세요.
###################################################

plt.rcParams["figure.figsize"] = (8,5)

# 클러스터링 과정을 5회 반복한다.
for i in range(5):
    # faith에 저장된 272개의 점에 대해 각각 점이 클러스터 0인지 클러스터 1인지를 기록하는 2차원 어레이
    # 아래처럼 272행 k(클러스터 수)열인 2차원 어레이
    #
    # rnk 어레이는 다음처럼 행번호에 해당하는 점이 열번호에 해당하는 묶음에 할당되면 1, 아니면 0으로 표시한다.
    #
    # rnk = [ [1, 0],      첫번째 점은 0번 묶음이다.
    #         [0, 1],      두번째 점은 1번 묶음이다.
    #         ...   ,
    #         ...   ,
    #         [1, 0] ]     272번째 점은 0번 묶음이다.
    #
    # 처음에는 모두 0으로 초기화
    rnk = np.zeros( (pass) )
    
    # 1) 모든 데이터에 대해서 초기 중심점과의 간격을 계산하여 더 가까운쪽의 묶음으로 데이터를 할당한다.
    # 각점이 어느 클러스터에 속하는지 rnk 어레이에 표시한다.
    for n in range(faith.shape[0]):
        # n번째 데이터 포인트를 x에 받아와서
        x = faith[n]
        
        # 현재 중심점 2개 centroids[0]와 centroids[1]간의 거리를 구해서 비교한다.
        # 초기 중심점은 다음처럼 윗셀에서 초기화 했다.
        # centroids = np.array([[-1.3, 1.0], [1.2, -1.0]])
        # 각 중심점에서 x까지 거리를 구하는 법 : np.linalg.norm(x-centroids[0]), np.linalg.norm(x-centroids[1])
        if pass : 
            rnk[n,0] = 1
        else :
            rnk[n,1] = 1
    
    # 그림을 그리기위해 faith에 저장된 점데이터를 파랑과 빨강으로 분류 한다.
    # 앞서 만든 rnk 어레이의 상태에 따라 두 부분으로 나눈다.
    blue = faith[np.where(rnk[:,0]==1)]
    red  = faith[np.where(rnk[:,1]==1)]
    
    ##########################################################################################
    # 2) 위 작업이 완료되면 두 묶음으로 분리된 데이터의 평균점을 구한다.
    # 힌트 : 지금 두 묶음은 위의 blue, red 어레이에 각각 들어가 있고
    # blue 어레이에 저장된 점의 평균을 centroids[0]에 할당하고
    # red 어레이에 저장된 점의 평균을 centroids[1]에 할당한다.
    # 어레이의 평균을 구하기 위해서는 mean 함수를 사용한다.
    # mean함수 문서 : https://docs.scipy.org/doc/numpy-1.14.0/reference/generated/numpy.mean.html
    centroids[0] = pass
    centroids[1] = pass
    
    # 한번 반복 후 클러스터링 상황을 그린다.
    plt.plot(blue[:,0], blue[:,1],'o', markersize=7, alpha=0.7)
    plt.plot(red[:,0],  red[:,1], 'o', markersize=7, alpha=0.7) 
    plt.plot(centroids[0,0], centroids[0,1], 'bx', markersize=10, markeredgewidth=5)
    plt.plot(centroids[1,0], centroids[1,1], 'rx', markersize=10, markeredgewidth=5)   

    plt.show()
````
