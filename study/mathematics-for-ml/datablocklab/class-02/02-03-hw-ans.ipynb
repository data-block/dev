### 2. 곱의 미분

$$
y = (2x^2 -2)(3x - x^3) \quad x \in [-2,2]
$$

- 다항식 미분 공식과 곱의 미분법을 써서 도함수를 구해보고 SymPy 사용하여 결과를 확인하고, 함수 그래프와 도함수의 그래프를 matplotlib를 사용하여 그려보세요.

- 곱의 미분을 이용한 미분 과정 

$$
\begin{align}
\frac{dy}{dx} &=(2x^2 -2)'(3x - x^3) + (2x^2 -2)(3x - x^3)' \\[5pt] 
&= (4x)(3x - x^3) + (2x^2 -2)(3-3x^2) \\[5pt]
&= 12x^2 -4x^4 + 6x^2 - 6x^4 - 6 + 6x^2 \\[5pt]
&= -10x^4 + 24x^2 -6
\end{align}
$$

- 아래 sympy 코드에서 pass로 표시된 부분을 적절한 표현으로 완성하시오.

````python
x = sympy.Symbol('x')

# diff 함수안에 함수를 표현하는 알맞은 코드를 적는다.
# diff함수 사용법 : sympy.diff( 함수표현, 미분할 변수 )
# pass 가 있는 위치에 알맞은 표현을 써넣는다.
sympy.simplify( sympy.diff((2*x**2 - 2)*(3*x - x**3) , x) )
````

- 아래 함수의 그래프와 도함수의 그래프를 그리는 코드에서 pass로 표시된 부분을 적절한 표현으로 완성하시오.

````python
plt.rcParams["figure.figsize"] = (5,10)

x = np.linspace(-2, 2, 100)

# 주어진 함수 y와 도함수 dy를 알맞게 초기화 한다.
# 변수는 위에서 생성한 x로 쓴다.
y  = (2*x**2 - 2)*(3*x - x**3) # 여기 주어진 함수 y의 알맞은 표현을 쓴다.
dy = -10*x**4 + 24*x**2 - 6    # 여기 주어진 함수의 도함수의 알맞은 표현을 쓴다.

fig = plt.figure()

ax1 = fig.add_subplot(2,1,1)
ax2 = fig.add_subplot(2,1,2)

ax1.plot(x, y, lw=3)
ax1.set_title(r"$f(x) = (2x^2 -2)(3x - x^3)$")

ax2.plot(x, dy, lw=3, color=style_colors[1])
ax2.set_title(r"Derivative of $f(x)$")
plt.show()
````

### 3. 분수함수 미분

$$
y = \frac{1}{x - 2}
$$

분수함수 미분을 이용하여 각각 편도함수를 구해보고 SymPy 사용하여 결과를 확인하세요.

- 분수함수 미분을 이용한 미분과정

$$
\frac{d }{dx} \left( \frac{1}{x - 2} \right) = \frac{1' \times (x-2) - 1 \times (x-2)'}{(x - 2)^2} = \frac{-1}{(x - 2)^{2} }
$$

- 합성함수 미분을 이용한 미분 과정

$$
\frac{d }{dx} \left( \frac{1}{x - 2} \right) = \frac{d }{dx}  (x - 2)^{-1} = -(x - 2)^{-2} \frac{d}{dx} (x - 2) = \frac{-1}{(x - 2)^{2} }
$$

- 아래 sympy 코드에서 pass로 표시된 부분을 적절한 표현으로 완성하시오.

````python
x = sympy.Symbol('x')

# diff 함수안에 함수를 표현하는 알맞은 코드를 적는다.
# diff함수 사용법 : sympy.diff( 함수표현, 미분할 변수 )
# pass 가 있는 위치에 알맞은 표현을 써넣는다.
sympy.diff(1 / (x - 2), x)
````

### 4. 다변수함수 미분

$$
f(x,y) = \frac{1}{2} x^2 + xy^3
$$

다항식 미분 공식을 써서 편도함수를 구해보고 SymPy 사용하여 결과를 확인하세요. 함수의 그래프와 각 편도함수의 그래프를 matplotlib를 사용하여 그려보세요.

- 편도함수 

$$
\begin{align}
& \frac{\partial}{\partial x} f(x,y) = x + y^3 \\[5pt]
& \frac{\partial}{\partial y} f(x,y) = 3xy^2
\end{align}
$$

- 아래 sympy 코드에서 pass로 표시된 부분을 적절한 표현으로 완성하시오.



````python
x = sympy.Symbol('x')
y = sympy.Symbol('y')

# diff 함수안에 함수를 표현하는 알맞은 코드를 적는다.
# diff함수 사용법 : sympy.diff( 함수표현, 미분할 변수 )
# pass 가 있는 위치에 알맞은 표현을 써넣는다.
sympy.diff((1/2)*x**2 + x*y**3, x)
````

````python
# diff 함수안에 함수를 표현하는 알맞은 코드를 적는다.
# diff함수 사용법 : sympy.diff( 함수표현, 미분할 변수 )
# pass 가 있는 위치에 알맞은 표현을 써넣는다.
sympy.diff((1/2)*x**2 + x*y**3, y)
````

- 아래 함수의 그래프와 도함수의 그래프를 그리는 코드에서 pass로 표시된 부분을 적절한 표현으로 완성하시오.

````python
plt.rcParams["figure.figsize"] = (17, 5)

from mpl_toolkits import mplot3d

x = np.linspace(-3, 3, 150)
xx, yy = np.meshgrid(x, x)

# 주어진 함수 f와 도함수 dfx, dfy를 알맞게 초기화 한다.
# 2변수 함수이기 때문에 변수는 위에서 생성한 메쉬그리드 변수 xx, yy로 쓴다.
f = (1/2)*xx**2 + xx*yy**3  # 여기 주어진 함수 f의 알맞은 표현을 쓴다.
dfx = xx + yy**3            # 여기 주어진 함수 dfx의 알맞은 표현을 쓴다.
dfy = 3*xx*yy**2            # 여기 주어진 함수 dfy의 알맞은 표현을 쓴다.

fig = plt.figure()
ax1 = fig.add_subplot(1, 3, 1, projection='3d')
ax2 = fig.add_subplot(1, 3, 2, projection='3d')
ax3 = fig.add_subplot(1, 3, 3, projection='3d')

ax1.plot_surface(xx, yy, f, rstride=10, cstride=10, cmap=plt.cm.viridis, edgecolor="k")
ax1.set_xlabel('x')
ax1.set_ylabel('y')
ax1.set_title(r'$\frac{1}{2} x^2 + xy^3$', fontsize=20)

ax2.plot_surface(xx, yy, dfx, rstride=10, cstride=10, cmap=plt.cm.viridis, edgecolor="k")
ax2.set_xlabel('x')
ax2.set_ylabel('y')
ax2.set_title(r'$\frac{\partial}{\partial x}\left( \frac{1}{2} x^2 + xy^3 \right)$', fontsize=20)

ax3.plot_surface(xx, yy, dfy, rstride=10, cstride=10, cmap=plt.cm.viridis, edgecolor="k")
ax3.set_xlabel('x')
ax3.set_ylabel('y')
ax3.set_title(r'$\frac{\partial}{\partial y}\left( \frac{1}{2} x^2 + xy^3 \right)$', fontsize=20)

plt.show()
````

### 5. 자동미분

다음 함수에 대해서 sympy를 사용하여 도함수를 구하고, 수치 미분과 자동 미분의 결과를 비교해보세요.

- 다음은 주어진 함수와 위 셀에서 sympy로 미분한 도함수의 그래프를 그린 것이다.
- 아래 함수의 그래프와 도함수의 그래프를 그리는 코드에서 pass로 표시된 부분을 적절한 표현으로 완성하시오.

````python
plt.rcParams["figure.figsize"] = (10, 5)

# 미분할 위치를 정하는 변수
x0 = 1

x = np.linspace(0, 3, 100)

# 주어진 함수 y와 도함수 dy를 함수 형태로 초기화 한다.
def y(x):
    """주어진 함수를 계산하는 함수"""
    return x / ((1 + x))      # 여기 주어진 함수 y의 알맞은 표현을 쓴다.

def dy(x):
    """주어진 함수 y의 도함수를 계산하는 함수"""
    return 1 / ( ((1+x)**2))  # 여기 주어진 함수 dy의 알맞은 표현을 쓴다.

plt.plot(x, y(x), label='y')
plt.plot(x, dy(x), label='dy')
plt.legend(fontsize=15)

dy0 = dy(x0)
plt.plot(x0, dy0, '.', markersize=15, color=style_colors[3])
plt.plot(x0, dy0, 'o', fillstyle='none', markersize=15, color=style_colors[3])
plt.text(x0, dy0+0.05, "derivative at x=1 : {:.2f}".format(dy0), fontsize=15)
plt.show()
````

- 그래프가 제대로 그려졌다면 $x=1$일때 도함수의 값이 0.25로 그려진다.

#### 수치미분

- 아래 코드는 주이전 함수를 전방 차분으로 수치미분하기 위한 코드이다.

$$
\frac{\partial \, f}{\partial \, x_i} \approx \frac{f(x_1, \cdots, \color{RoyalBlue}{x_i + \Delta x_i} , \cdots, x_n) - f(x_1, \cdots, x_i, \cdots, x_n)}{\Delta x_i}
$$

- sympy로 계산한 미분계수 0.25 근처 값이 출력됨을 확인할 수 있다.



````python
def numer_deriv(f, x, eps=0.01):
    """
    f   : 미분할 함수
    x   : 미분계수를 구할 독립변수 값
    eps : 차분을 구하기 위한 미소변량
    """
    diff = ( f(x+eps) - f(x) ) / eps
    
    return diff

print( "derivative at x=1 : {:.2f}".format(numer_deriv(y, x0)) )
````

#### 자동미분

- 아래 코드는 주어진 함수를 자동미분하기 위해 필요한 게이트를 클래스로 구한한 코드이다.

````python
###############################################################
# 주어진 함수를 그래프로 만들기 위해 필요한 게이트 클래스 구현
# ctrl+enter를 눌러서 셀을 한번 실행시켜야 함
###############################################################

class Multiply :
    """
    https://github.com/WegraLee/deep-learning-from-scratch
    """
    def __init__(self):
        self.x = None
        self.y = None
        
    def __repr__(self):
        return "Mutiply"
        
    def forward(self, x, y):
        self.x = x
        self.y = y    
        
        return x * y

    def backward(self, dout):
        dx = dout * self.y  # x와 y를 바꾼다.
        dy = dout * self.x

        return dx, dy


class Add :
    """
    https://github.com/WegraLee/deep-learning-from-scratch
    """
    def __init__(self):
        pass
    
    def __repr__(self):
        return "Add"
    
    def forward(self, x, y):
        return x + y

    def backward(self, dout):
        dx = dout * 1
        dy = dout * 1

        return dx, dy
    
class Inverse :
    def __init__(self):
        self.x = None
        
    def forward(self, x) :
        self.x = x
        
        return 1./x
    
    def backward(self, dout):
        return -(1 / self.x**2)*dout
````

- 주어진 함수의 그래프 표현은 위와 같고 이를 순전파, 역전파 시키기 위한 함수는 아래와 같다.

- 순전파를 위해 a, b, y 순서로 진행시킨다.

```python
a_fwd = a.forward(x, 1)
b_fwd = b.forward(a_fwd)
forward = y.forward(x, b_fwd)
```

- 순전파 결과를 이용해서 역전파 시킨다. y에 대해 1을 넘기면서 역전파를 시작한다.

```python
y_bak = y.backward(1)
```

- a와 b를 순서대로 역전파시킬 때 인자로 어떤 값을 넘겨야 하는지 pass부분을 채워 완성하시오. 

- 결과는 다음처럼 출력됨 : Func. val. by forward pass = 0.50, Derivative by backward pass = 0.25

````python
# 아래 함수에서 pass로 된 부분에 적당한 코드를 채워넣으세요.

def autodiff(x) :
    # 각 게이트를 정의 한다.
    # 위 그래프 그림에서 a는 x와 1을 입력으로 받는 더하기(Add) 함수
    a = Add()
    
    # b는 입력에 역수를 취하는 역수(Inverse) 함수
    b = Inverse()
    
    #y 는 x와 b를 입력으로 받는 곱하기(Multiply) 함수
    y = Multiply()
    
    ######################################################################
    #forward
    # 게이트 순서대로 전진한다.
    a_fwd = a.forward(x, 1)
    b_fwd = b.forward(a_fwd)
    forward = y.forward(x, b_fwd)
    ######################################################################
    
    ######################################################################
    #backward
    # 게이트의 역순으로 후진한다.
    y_bak = y.backward(1)
    
    # b함수와 a함수 순으로 후진할때 각각 backward함수에 어떤 인자를 
    # 넘겨야 하는가?
    b_bak = b.backward(y_bak[1])
    a_bak = a.backward(b_bak)
    ######################################################################
    
    # x 노드에 해당하는 backward 결과를 다 더한다.
    backward = y_bak[0] + a_bak[0]
    
    return forward, backward

print("Func. val. by forward pass = {:.2f}, Derivative by backward pass = {:.2f}".format(*autodiff(x0)) )
````
